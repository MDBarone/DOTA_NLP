```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import ExcelWriter

df = pd.read_csv('/Users/arayi/Downloads/Momentum Factor Testing Data.csv')

df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Date'] = pd.DatetimeIndex(df['Date'])
df.set_index('Date', inplace=True)

# creating a dataframe with the daily pct_change for each of the stocks in the initial dataframe
df_day_change = df.pct_change(-1)

# Number of trading days we are using to measure momentum
mom_period = 252

# This is the dataframe that converts the stock price data into 6 month percent change
df_mom = (df.pct_change(mom_period)).round(3)

# this produces a dataframe where the stock names are returned for a long or short position
# depending on their recent 6 month momentum - top 3 are long, bottom 3 are short
df_3max = pd.DataFrame({'Long Stocks': df_mom.apply(lambda s: s.nlargest(3).index.tolist(), axis=1),
                        'Short Stocks': df_mom.apply(lambda s: s.nsmallest(3).index.tolist(), axis=1)})

# this removes the rows where the value of Long stock & short stock columns is an empty list
df_3max.drop(df_3max.index[:mom_period], inplace=True)

print(df_3max.head())

# creating the columns to populate with daily portfolio returns
df_3max['Long Perf'] = None
df_3max['Short Perf'] = None

long_list = []
short_list = []

for row in range(len(df_3max)):

    for i in df_3max['Long Stocks'].iloc[row]:
        long_item = df_day_change[i][row + mom_period]
        long_list.append(long_item)

    for i in df_3max['Short Stocks'].iloc[row]:
        short_item = df_day_change[i][row + mom_period]
        short_list.append(short_item)

x = 0
y = 0

for row in range(len(df_3max)):

    num_long = len(df_3max['Long Stocks'].iloc[0])
    num_short = len(df_3max['Short Stocks'].iloc[0])

    df_3max['Long Perf'][row] = (sum(long_list[x:num_long + x])) / num_long
    df_3max['Short Perf'][row] = (sum(short_list[y:num_short + y])) / num_short

    x += num_long
    y += num_short

df_3max['L/S Perf'] = (df_3max['Long Perf'] - df_3max['Short Perf']) / 2


((df_3max['Long Perf'] + 1).cumprod()).plot()
((df_3max['Short Perf'] + 1).cumprod()).plot()
((df_3max['L/S Perf'] + 1).cumprod()).plot()
plt.legend()
plt.show()

# writer1 = ExcelWriter('ExportTEST2.xlsx')
# df_3max.to_excel(writer1, 'Sheet2')
# writer1.save()

# writer2 = ExcelWriter('ExportTEST2.xlsx')
# df_day_change.to_excel(writer2, 'Sheet2')
# writer2.save()
```
